```python
import os, sys
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
#!pip3 install --upgrade seaborn # Updates the seaborn version in the collab notebook. if needed

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.utils import check_random_state
from sklearn import datasets
from sklearn.manifold import TSNE
from time import time

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

```

```python
df_wine = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data',header=None)
df_wine.head(5)
```

```python
df_wine.columns = ['Class label', 'Alcohol', 'Malic acid', 'Ash',
                   'Alcalinity of ash', 'Magnesium', 'Total phenols',
                   'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins',
                   'Color intensity', 'Hue', 'OD280/OD315 of diluted wines',
                   'Proline']
df_wine.head(1)
```

```python
df_wine['Class label'].unique()
```

```python
y=df_wine['Class label']
y.unique()
```

```python
X = df_wine.drop('Class label',1)  # Created a sub-DataFrame, drop 'Class Label'
print(type(X))
X.head(4)
#X.shape
```

```python
'''standardizing the data'''
sc = StandardScaler()          # mu and sigma is for Gaussian distribution/Normal dist.
sc.fit(X)
# sc.fit() will estimate the parmeters: mu and sigma for each feature dimention of the data.

X_std = sc.transform(X)
# sc.transform() will standardize the data, based on estimated parameters from sc.fit().

#print(X_train_std[:])
X_std.shape
```

## TSNE

#### TSNE with perplexity at 5 and learning rate at 100.

```python
tsne = TSNE(n_components=2,learning_rate=100,perplexity=20,random_state=0)
X_new=tsne.fit_transform(X_std)
```

```python
df = pd.DataFrame()

df["tsne-1"] = X_new[:,0]
df["tsne-2"] = X_new[:,1]
df["y"] = y

sns.scatterplot(x="tsne-1", y="tsne-2", hue=df.y.tolist(),
                palette=sns.color_palette("hls", 3),
                data=df).set(title="Wine: perp 20, lr 100, iter 1,000",alpha=0.6)
```

```python
tsne = TSNE(n_components=2,learning_rate=150,perplexity=20,random_state=0,n_iter=2000)
X_new=tsne.fit_transform(X_std)

sns.scatterplot(x="tsne-1", y="tsne-2", hue=df.y.tolist(),
                palette=sns.color_palette("hls", 3),
                data=df).set(title="Wine: perp 20, lr 150, iter 2,000",alpha=0.6)
```
